You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:

        10

       /   \

     2      7

   /   \

 8      4

Output:

        8

      /   \

    4     10

  /   \

2      7

</aside>

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def binary_tree_to_bst(root):
    # Perform in-order traversal to get sorted values
    def inorder_traversal(node):
        if node is None:
            return []
        return inorder_traversal(node.left) + [node.val] + inorder_traversal(node.right)

    def build_bst(values):
        if not values:
            return None

        # Find middle element as the root
        mid = len(values) // 2
        root = TreeNode(values[mid])

        # Recursively build left and right subtrees
        root.left = build_bst(values[:mid])
        root.right = build_bst(values[mid + 1:])

        return root

    # Get sorted values from in-order traversal
    values = inorder_traversal(root)

    # Build binary search tree using sorted values
    bst_root = build_bst(values)

    return bst_root


# Create the binary tree
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Convert binary tree to binary search tree
bst_root = binary_tree_to_bst(root)

# Function to print in-order traversal of a binary tree
def print_inorder(node):
    if node is None:
        return
    print_inorder(node.left)
    print(node.val, end=" ")
    print_inorder(node.right)

# Print the in-order traversal of the resulting binary search tree
print_inorder(bst_root)
```

Write a program to convert a binary tree to a doubly linked list.

Input:

        10

       /   \

     5     20

           /   \

        30     35

Output:

5 10 30 20 35

</aside>

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def binary_tree_to_doubly_linked_list(root):
    def convert_to_doubly_linked_list(node):
        nonlocal prev

        if node is None:
            return

        convert_to_doubly_linked_list(node.left)

        node.left = prev
        if prev:
            prev.right = node

        prev = node

        convert_to_doubly_linked_list(node.right)

    prev = None
    convert_to_doubly_linked_list(root)

    # Traverse the doubly linked list
    while root and root.left:
        root = root.left

    # Print the values of the doubly linked list
    while root:
        print(root.val, end=" ")
        root = root.right


# Create the binary tree
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

# Convert binary tree to doubly linked list and print the values
binary_tree_to_doubly_linked_list(root)
```
